```python
#learning basic numpy
import numpy as np
x = np.array([[1,2,3,7,],[4,5,6,8]])
print("x:\n{}".format(x))
```

```python
#learning basic SciPy
from scipy import sparse

#creat a 2D Numpy array with a diagonal of ones, and zeros everywhere else
eye = np.eye(4)
print("NumPy Array:\n{}".format(eye))
```

```python
#convert the numpy array to a scipy sparse matrix in CSR Format
#only the nonzero entries are stored

sparse_matrix = sparse.csr_matrix(eye)
print("\n SciPy sparse CSR matrix:\n{}". format(sparse_matrix))
```

```python
#learn basic matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

#generate a sequence of number from -35, 0, 35 with 100 steps in between
x = np.linspace(-35,0,35)

#create a second array using sine
y = np.sin(x)

#the plot function makes a line chart of one array against another
plt.plot(x,y, marker="x")
```
